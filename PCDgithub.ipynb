```python
import cv2
```

```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
img = cv2.imread(r"C:\Users\USER PC\medi.jpg")
```

```python
plt.imshow(img)
```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gray, cmap = 'gray')
```

```python
img_copy = img.copy()
```

```python
cv2.rectangle(img_copy, pt1 = (220, 900), pt2 = (400, 1222), color = (255, 0, 0), thickness = 5)
plt.imshow(img_copy)
```

```python
cv2.circle(img, center = (1500, 400), radius = 100, color = (0, 0, 255), thickness = 8)
plt.imshow(img)
```

```python
cv2.putText(img, text = "My Image", org = (220, 400), fontFace = cv2.FONT_HERSHEY_DUPLEX, fontScale = 5,
            color = (0, 255, 0), thickness = 10, lineType = cv2.LINE_AA)
plt.imshow(img)
```
